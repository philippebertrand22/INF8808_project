```python
import kagglehub
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

path = kagglehub.dataset_download("yamaerenay/spotify-dataset-19212020-600k-tracks")

print("Path to dataset files:", path)
df = pd.read_csv(path+"/tracks.csv")
```

```python

def convert_dates(dataframe):
    dataframe["release_date"] = pd.to_datetime(dataframe["release_date"],errors = "coerce")   
    return dataframe

def filter_years(dataframe, start, end):
    mask = (dataframe["release_date"].dt.year >= start) & (dataframe["release_date"].dt.year <= end)
    dataframe = dataframe.loc[mask]
    return dataframe
def summarize_yearly_counts(dataframe):
    dataframe["Year"] = dataframe["release_date"].dt.year
    return dataframe.sort_values(["popularity","Year"],ascending = False).reset_index(drop = True)

df = convert_dates(df)
df = filter_years(df,1922,2020) #can change this range
df = summarize_yearly_counts(df)
print(df.columns)
print(df.shape)
df.head()
```

```python
df_sub = df[['popularity', 'duration_ms','danceability', 'energy','loudness', 'speechiness', 'valence', 'tempo', 'Year']]
df_n = df_sub.groupby("Year",as_index = False).mean() #mean of yearly mean
df_n
```

```python
fig = px.line(df_n, x = "Year",y = "popularity")
fig
```

```python
from jinja2 import Template
```

```python
input_html = "./template.html"
output_html = "./index.html"
```

```python
plotly_fig = {"fig":fig.to_html(full_html=False)}

with open(output_html, "w", encoding="utf-8") as output_file:
    with open(input_html) as template:
        j2_template = Template(template.read())
        output_file.write(j2_template.render(plotly_fig))
```
